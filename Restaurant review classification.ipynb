```python
# Creating dataframe
import pandas as pd
reviews_df=pd.read_csv('/content/Restaurant_Reviews.tsv',delimiter='\t')
reviews_df
```

```python
#Processing data
reviews_df.shape
```

```python
reviews_df.info()
```

```python
reviews_df['Liked'].value_counts()#1-positive review and 0-Negative review
```

**DATA VISUALIZATION**

```python
import seaborn as sns
sns.countplot(x=reviews_df.Liked)
```

```python
x=reviews_df['Review'].values 
y=reviews_df['Liked'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

**COUNT VECTORIZER**

```python
# count vectorization / Bag of Words technique
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
x_train_vect=vect.fit_transform(x_train)
x_test_vect=vect.transform(x_test)
```

**SUPPORT VECTOR MACHINES MODEL 1**

```python
from sklearn.svm import SVC
model1=SVC()
```

```python
model1.fit(x_train_vect,y_train)
```

```python
y_pred1=model1.predict(x_test_vect)
y_pred1 #predicted output
```

```python
y_test  #actual output
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1,y_test)
```

**PIPELINE WITH VECTORIZATION MODEL FOR COUNT VECTORIZER AND SVC MODEL 2**

```python
from sklearn.pipeline import make_pipeline
model2 = make_pipeline(CountVectorizer(),SVC())

model2.fit(x_train,y_train)
y_pred2 = model2.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred2,y_test)
```

**MULTINOMIAL NAIVE BAYES MODEL 3**

```python
from sklearn.naive_bayes import MultinomialNB
model3 = MultinomialNB()
```

```python
model3.fit(x_train_vect,y_train)
```

```python
y_pred3 = model3.predict(x_test_vect)
y_pred3 
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred3,y_test)
```

**PIPELINE WITH VECTORIZATION MODEL FOR COUNT VECTORIZER AND MULTINOMIAL NAIVE BAYES CLASSIFIER MODEL 4**

```python
from sklearn.pipeline import make_pipeline 
model4 = make_pipeline(CountVectorizer(),MultinomialNB())

model4.fit(x_train,y_train)
y_pred4 = model4.predict(x_test)
y_pred4
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred4,y_test)
```

```python
#COMPARING ACCURACY SCORES OF FOUR MODELS


# model1 accuracy - 0.72
# model2 accuracy - 0.792
# model3 accuracy - 0.744
# model4 accuracy - 0.784 
```

**Sentiment analysis**

*Sentiment analysis is the automated classification of reviews i.e. 
positive,negative and neutral.

*It uses Natural Language Processing(NLP) which is subset of AI.

*It is used to understand emotions of the writer about a product or service.

*To implement this, huge data and machine learning models like countvectorizer,  SVC, MultiNomial Naive Bayes etc. are required to train the data and to be used for classification.

*Social Media giants use this analysis for marketing or reviewing their product or services.

**Twitter Sentiments**

*Twitter uses sentiment analysis to differentiate positive,negative,and neutral tweets.

*This also helps businesses in campaigning, product launch advertising and reviewing on twitter.

*Sometimes twitter sentiments are used for exit polls.

*Twitter acts as platform for public opinions.

*It also helps in expressing common emotions through trends.

*To achieve this twitter developed NLP algorithm using Naive Bayes,SVM etc.
